### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí

# Promedio

sum_hope = 0
for i in range(len(esperanzas_de_vida)):
    sum_hope += esperanzas_de_vida[i]
promedio = sum_hope / len(esperanzas_de_vida)
print(f"El promedio de la esperanza de vida de los 10 países es de {promedio:.2f} años")

# Máximo
maximo = esperanzas_de_vida[0]
for i in range(len(esperanzas_de_vida)):
    if esperanzas_de_vida[i] > maximo:
        maximo = esperanzas_de_vida[i]
    else:
        continue
print(f"El máximo es {maximo}")

# Mínimo
minimo = esperanzas_de_vida[0]
for i in range(len(esperanzas_de_vida)):
    if esperanzas_de_vida[i] < minimo:
        minimo = esperanzas_de_vida[i]
    else:
        continue
print(f"El mínimo es {minimo}")
````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí

def clasificar_esperanza_vida(years):
    if years >= 75:
        return "Alta"
    elif years > 60 and years <= 74.9:
        return "Media"
    else:
        return "Baja"
        
for clasificacion in [86, 75, 65, 74.9, 60, 54]:
    print(f"Años = {clasificacion} => {clasificar_esperanza_vida(clasificacion)}") 
````

*¿Por qué usamos la media para rellenar nulos en vez de eliminar las filas?*

> Tu respuesta:
Si eliminamos toda la fila donde encontremos datos nulos, estaríamos obviando información importante de otros campos que harían que nuestro dataset cambiara completamente reduciendo considerablemnte la cantidad de datos disponibles y afectando la calidad del análisis. Para evitar esto se usa la media, para rellenar los valores nulos y así evitamos perder información, manteniendo la estructura del dataset.
